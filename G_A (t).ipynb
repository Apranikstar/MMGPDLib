```python
import MMGPD
import numpy as np
import matplotlib.pyplot as plt
```

```python
GA = MMGPD.G_A(0.1, 1e-1) # first argument is t and second is the percision of integration
```

```python
GA
```

```python

t_values = np.linspace(-5, -0.1, 15)  # 100 points between -5 and -0.1

# Compute the integrals for each value of t

GAT = [MMGPD.G_A(t,1e-1)[0] for t in t_values]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(-t_values, GAT, color='blue', label='Results')
plt.plot(-t_values, GAT, color='lightblue', alpha=0.5)  # Optional line connecting the points
plt.title('G_A(t) vs. -t Values')
plt.xscale('log')
plt.xlabel('t Values')
plt.yticks(np.linspace(0, 1.5, 15))
plt.ylabel('G_A(t)')
plt.legend()
plt.grid()
plt.show()
```
